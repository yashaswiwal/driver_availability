```python
df = pd.merge(df,bdays, left_on = 'driver',right_on = 'DRVR_USER_ID')
```

```python
df = df.drop(columns = ['MARKET_NM','DRVR_USER_ID'], axis = 1)
```

```python
df['DRVR_BIRTH_DT'] = pd.to_datetime(df['DRVR_BIRTH_DT'])
```

```python
def age(birthdate):
    today = datetime.today()
    age = today.year - birthdate.year
    return age
```

```python
df['age'] = df['DRVR_BIRTH_DT'].apply(lambda x: age(x))
```

```python
df = df.drop('DRVR_BIRTH_DT',axis=1)
```

```python
df = df.drop_duplicates()
```

```python
df = df.reset_index(drop=True)
```
